```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn.linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("bigcity (1).csv")


print("data description",data.describe())
missing_count = data.isna().sum()
print("number of missing values in each column\n",missing_count)
print("data shape", data.shape)
```

```python
x=data.iloc[:, :-1].values
y=data.iloc[:,1].values

print("x shape", x.shape)
```

```python
corr_matrix=data.corr()

plt.figure(figsize=(4, 4))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=1)


print("X_train shape:", x_train.shape)
print("y_train shape:", len(y_train))  
print("X_test shape:", x_test.shape)
print("y_test shape:", len(y_test))    
```

```python
regression_model = LinearRegression()
regression_model.fit(x_train, y_train)

coefficients=regression_model.coef_
print("coefficients",coefficients)

intercept=regression_model.intercept_
print("intercept", intercept)
```

```python
y_pred = regression_model.predict(x_test)
```

```python


plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regression_model.predict(x_train), color = 'blue')
plt.title('Population: 1920 vs 1930 (Training set)')
plt.xlabel("Population (in 1000's) in 1920.")
plt.ylabel("Population (in 1000's) in 1930.")
plt.show()
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
print(len(y_pred))
print(len(y_test))
print(len(y))



rmse = mean_squared_error(y_test, y_pred)**0.5
r2 = r2_score(y_test, y_pred)
print("Root Mean Square Error: ", rmse)
print("R2 Score:", r2)


```
